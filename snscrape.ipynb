```python
import re
# function for twitter text cleaning
def cleanTwt(text):
    #remove \n
    text = re.sub("\\\\n", " ", text)
    #remove #
    text = re.sub("#", "", text)
    #remove twitter handles (@user)
    text = re.sub("(@[A-Za-z0-9_]+)","", text)
    #remove url links and 
    text = re.sub("https?://[A-Za-z0-9./]*","",text)
    # remove whitespaces
    text = ' '.join(text.split())
    
    return text

```

```python
df['Cleaned_Tweets'] = df['content'].apply(lambda x:cleanTwt(x))
df.head()
```

```python
#downloading the vader lexicon from nltk
import nltk
nltk.download('vader_lexicon')
#importing sentimentintensityanaylzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
df['Scores'] = df['Cleaned_Tweets'].apply(lambda text: sid.polarity_scores(text))
df.head(100)
```

```python
df['Scores'] = df['Scores'].astype(str)
```

```python
#extract the compound score from substring
df['New']= df['Scores'].str.split(':').str[4]
df['Sentiment']= df['New'].str.split('}').str[0]
```

```python
df= df.drop(['New'], axis=1)
df
```

```python
df = df.drop(df.columns[0], axis=1)
df
```

```python
df['Sentiment'] = df['Sentiment'].astype(float)
df
```
